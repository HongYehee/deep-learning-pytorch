### 국민청원 분류하기

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# load crawling.csv
df = pd.read_csv('data/crawling.csv') # thousands=True => 1,000단위 콤마 제거 후 숫자로 가져옴
df.head()
```

#### pre-processing

```python
# before preprocessing (index:1, col:content)
df.loc[1]['content']
```

```python
# regular expression
import re

def remove_white_space(text):
    text = re.sub(r'[\t\r\n\f\v]', ' ', str(text))
    return text

def remove_spacial_char(text):
    # ㄱ~ㅣ, 가~힣, 0 ~ 9, A ~ Z, a ~ z에 해당하지 않는 것
    text = re.sub('[^ㄱ-ㅣ가-힣0-9A-Za-z]+', ' ', str(text))
    return text
```

```python
# row: 1000까지
df = df.iloc[:1000].copy()
df.head()
```

```python
# title
df['title'] = df['title'].apply(remove_white_space)
df['title'] = df['title'].apply(remove_spacial_char)

# content
df['content'] = df['content'].apply(remove_white_space)
df['content'] = df['content'].apply(remove_spacial_char)
```

```python
# after preprocessing
df.loc[1]['content']
```

#### Tokenizing and create variable

```python
# konlpy for Korean text
from konlpy.tag import Okt, Hannanum, Kkma, Komoran
okt = Okt()
hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()

a1 = okt.morphs('동해물과 백두산이 마르고 닳도록 하느님이 보우하사 우리나라 만세')
a2 = hannanum.morphs('동해물과 백두산이 마르고 닳도록 하느님이 보우하사 우리나라 만세')
a3 = kkma.morphs('동해물과 백두산이 마르고 닳도록 하느님이 보우하사 우리나라 만세')
a4 = komoran.morphs('동해물과 백두산이 마르고 닳도록 하느님이 보우하사 우리나라 만세')
```

```python
print(a1)
print(a2)
print(a3)
print(a4)
```

```python
# return (형태소, 품사)
okt.pos('동해물과 백두산이 마르고 닳도록 하느님이 보우하사 우리나라 만세')
```

```python
# return nouns
okt.nouns('동해물과 백두산이 마르고 닳도록 하느님이 보우하사 우리나라 만세')
```

```python
# make okt
okt = Okt()

# title tokenizing
df['title_token'] = df['title'].apply(okt.nouns)
df['content_token'] = df['content'].apply(okt.nouns)
df.head()
```

```python
# title token + content token
df['token_final'] = df['title_token'] + df['content_token']
df.head()
```

```python
# count: remove ',' and convert to int
df['count'] = df['count'].replace({',' : ''}, regex=True).apply(lambda x : int(x))
df.head()
```

```python
# make label col: over 1000 counts -> YES
df['label'] = df['count'].apply(lambda x : 'YES' if x >= 1000 else 'NO')
df.head()
```

```python
# extract token and label
df_drop = df[['token_final', 'label']]
df.head()
```

```python
# save
df_drop.to_csv('data/df_drop.csv', index=False, encoding='utf-8-sig')
```
