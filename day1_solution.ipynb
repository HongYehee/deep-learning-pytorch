2. x_data = [1,2,3,4,5]
y_data = [5,8,11,14,17]
인 경우 w, b를 구하시오.
1) 미분을 직접 구현
2) 파이토치를 이용하여

```python

def mse( x,y, w ,b):
    c = 0
    for i in range(len(x)):
        hx = w*x[i] +b
        c = c+(hx-y[i])**2
    return c/len(x)
```

```python
def gradientW( x,y, w, b ):
    c = 0
    for i in range(len(x)):
        hx = w*x[i]+b
        c = c+(hx-y[i])*x[i]*2
    return c/len(x)
```

```python
def gradientB( x,y, w ,b):
    c = 0
    for i in range(len(x)):
        hx = w*x[i]+b
        c = c+(hx-y[i])*2
    return c/len(x)
```

```python
w = 10
b= 10
x_data = [1,2,3,4,5]
y_data = [5,8,11,14,17]
for epoch in range(1000):
    c = mse( x_data, y_data, w,b )
    gw = gradientW( x_data, y_data, w ,b)
    gb = gradientB( x_data, y_data, w ,b)
    print( 'w=',w, 'cost=',c, 'gradient=', gw, gb )
    w = w-0.1*gw
    w = w-0.1*gb
print( '최종w=', w)
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np

x_data = [1,2,3,4,5]
y_data = [5,8,11,14,17]
x = torch.FloatTensor(x_data)
y = torch.FloatTensor(y_data)
#미분을 직접하는 경우:
epoch = 2000
learning_rate = 0.01
n= 5
w = torch.tensor(10.0, requires_grad=True)
b = torch.tensor(10.0, requires_grad=True)

for i in range(epoch):
    hx = w*x + b
    c= torch.sum((hx-y)**2)/n
    gradientW = torch.sum((w*x+b -y)*2*x)/n 
    gradientB = torch.sum((w*x+b-y)*2)/n
    
    w = w -learning_rate * gradientW
    b  = b - learning_rate * gradientB
    print('cost: ', c.item(), 'w=', w.detach().item(), 'b=', b.detach().item())

print('최종 w=',  w.detach().item(), '최종 b=', b.detach().item())   
```

```python
import torch
import torch.nn.functional as F
from  torch.optim  import Adam
import matplotlib.pyplot as plt
import numpy as np
x_data = torch.FloatTensor( [1,2,3,4,5])
y_data = torch.FloatTensor( [5,8,11,14,17])
w = torch.tensor( 10.0, requires_grad=True)
b = torch.tensor( 10.0, requires_grad=True)

optimizer = Adam( [w,b], lr=0.1  )
for i in range(3000):
    hx = w*x_data + b
    c = torch.mean( (hx-y_data)**2)
    optimizer.zero_grad() # 미분 초기화
    c.backward() # 미분( w, b )
    optimizer.step() # w= w-0.1* w미분값 b= b-0.1*b미분값
    print( i, 'cost:', c.item() )
print('최종cost:', c.item(), '최종w:', w.detach().item(), '최종b:', b.detach().item())
```
