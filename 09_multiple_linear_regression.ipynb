### Multiple Linear Regression
feature가 여러 개 있는 것  

예) y = w1x1 + w2x2 ⇒ 그래프를 그리면 3차원

```python
# 필요한 라이브러리 호출
import torch
import torch.nn.functional as F
from torch.nn import Linear, MSELoss, Sequential
from torch.optim import Adam
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# data : 공부 시간과 출석 일수에 따른 시험 점수
x1_data = [1, 2, 3, 4, 5] # 공부 시간
x2_data = [1, 2, 3, 4, 5] # 출석 일수
y_data = [50, 60, 70, 80, 90] # 시험 점수
```

```python
# 텐서 노드 만들기
x1 = torch.FloatTensor(x1_data)
x2 = torch.FloatTensor(x2_data)
y = torch.FloatTensor(y_data)

w1 = torch.empty(1, requires_grad=True)
w2 = torch.empty(1, requires_grad=True)
b = torch.empty(1, requires_grad=True)

torch.nn.init.uniform_(w1)
torch.nn.init.uniform_(w2)
torch.nn.init.uniform_(b)
```

```python
# cost function 정의
def cost():
    hx = w1 * x1 + w2 * x2 + b
    c = torch.mean((hx - y) ** 2)
    return c
```

```python
# 학습
optimizer = Adam([w1, w2, b], lr=0.01)

for epoch in range(1000):
    optimizer.zero_grad()
    c = cost()
    c.backward()
    optimizer.step()
    print(f'{epoch} - cost : {c.item()}')
```

```python
# 5시간 공부하고 5번 출석한 경우 점수 예측하기
w1 * 5 + w2 * 5 + b
```
