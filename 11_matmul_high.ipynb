### Multiple Linear Regression - matmul (high level)

```python
# 필요한 라이브러리 호출
import torch
import torch.nn.functional as F
from torch.nn import Linear, MSELoss, Sequential
from torch.optim import Adam
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# test.csv 불러오기
df = pd.read_csv('data/test.csv', header=None)
df.columns = ['q1', 'q2', 'mid', 'final']
df.head()
```

```python
# row는 전체, col은 마지막만 추출
x_data = df.iloc[:, :-1].values
y_data = df.iloc[:, [-1]].values
```

```python
# 텐서 노드 만들기 (w, b 정의 안 해도 됨)
x = torch.FloatTensor(x_data)
y = torch.FloatTensor(y_data)
```

```python
# 학습
model = Sequential()
model.add_module('nn1', Linear(3, 1))
loss_fn = MSELoss()
optimizer = Adam(model.parameters(), lr=0.1)

for epoch in range(2000):
    optimizer.zero_grad()
    hx = model(x)
    cost = loss_fn(hx, y)
    cost.backward()
    optimizer.step()
    print(f'{epoch} - cost : {cost.item()}')
```

```python
# 학습된 weight 값 확인
model[0].weight
```

```python
# 학습된 bias 값 확인
model[0].bias
```
