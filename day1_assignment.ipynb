#### 1) ÏßÅÏ†ë ÎØ∏Î∂Ñ

```python
epoch = 2000
learning_rate = 0.01
n = 5

x = torch.FloatTensor([1, 2, 3, 4, 5])
y = torch.FloatTensor([5, 8, 11, 14, 17])
w = torch.tensor(10.0, requires_grad=True)
b = torch.tensor(10.0, requires_grad=True)
```

```python
for i in range(epoch):
    hx = w * x + b
    c = torch.sum((hx - y) ** 2) / n
    gradient_w = torch.sum((w * x + b - y) * 2 * x) / n
    gradient_b = torch.sum((w * x + b - y) * 2) / n
    w = w - learning_rate * gradient_w
    b = b - learning_rate * gradient_b
    print(f'cost: {c.item()}, w: {w.detach().item()}, b: {b.detach().item()}')
    
print('=======================================================')
print(f'ÏµúÏ¢Ö w: {w.detach().item()}, ÏµúÏ¢Ö b: {b.detach().item()}')
print('=======================================================')
```

#### 2) PyTorch Ìï®Ïàò Ïù¥Ïö©

```python
x = torch.FloatTensor([1, 2, 3, 4, 5])
y = torch.FloatTensor([5, 8, 11, 14, 17])
w = torch.tensor(10.0, requires_grad=True)
b = torch.tensor(10.0, requires_grad=True)
```

```python
optimizer = Adam([w, b], lr=0.1)

for i in range(2000):
    hx = w * x + b
    c = torch.mean((hx - y) ** 2)
    optimizer.zero_grad()
    c.backward()
    optimizer.step()
    print(f'{i} - cost: {c.item()}')

print('=======================================================')
print(f'ÏµúÏ¢Ö w: {w.detach().item()}, ÏµúÏ¢Ö b: {b.detach().item()}')
print('=======================================================')
```
