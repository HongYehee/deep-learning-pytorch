### Multiple Linear Regression (matmul)

```python
# 필요한 라이브러리 호출
import torch
import torch.nn.functional as F
from torch.nn import Linear, MSELoss, Sequential
from torch.optim import Adam
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# test.csv 불러오기
df = pd.read_csv('data/test.csv', header=None)
df.columns = ['q1', 'q2', 'mid', 'final']
df.head()
```

```python
# row는 전체, col은 마지막만 추출
x_data = df.iloc[:, :-1].values
y_data = df.iloc[:, [-1]].values
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
# 텐서 노드 만들기 (행렬)
x = torch.FloatTensor(x_data) # x1과 x2를 같이
y = torch.FloatTensor(y_data)

w = torch.empty([3, 1], requires_grad=True) # [특성 데이터 수, 라벨 수]
b = torch.empty([1], requires_grad=True) # [라벨 수]

torch.nn.init.uniform_(w)
torch.nn.init.uniform_(b)
```

```python
# cost function 정의
def cost():
    hx = torch.matmul(x, w) + b
    c = torch.mean((hx - y) ** 2)
    return c
```

```python
# 학습
optimizer = Adam([w, b], lr=0.01)

for epoch in range(2000):
    optimizer.zero_grad()
    c = cost()
    c.backward()
    optimizer.step()
    print(f'{epoch} - cost : {c.item()}')
```

```python
# 예측 함수 정의
def hx_fx(xd):
    xd = torch.tensor(xd, dtype=torch.float32)
    hx = torch.matmul(xd, w) + b
    return hx.detach().numpy()
```

```python
# 예측해보기 : q1 80점, q2 90, mid 90점인 경우의 최종 점수 예측하기
hx_fx([[80, 90, 90]]) # 1행 3열로 주기
```
